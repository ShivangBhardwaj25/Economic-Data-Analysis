# ECONOMICS DATA ANALYSIS

### Objective Of The Project:

#### The objective of this project is to analyze global cost of living data to understand economic differences across countries.

#### This analysis focuses on identifying patterns and relationships between key economic indicators such as:
#### - Cost of Living Index
#### - Rent Index
#### - Groceries Index
#### - Restaurant Price Index
#### - Local Purchasing Power Index

### The Main Goals Of This Project Are:
#### - To identify the most and least expensive countries to live in
#### - To analyze how rent and groceries impact overall cost of living
#### - To study the relationship between cost of living and purchasing power
#### - To compare affordability across different countries
#### - To derive meaningful insights using data visualization techniques

#### This project uses Python libraries such as Pandas, NumPy, and Matplotlib for data cleaning, analysis, and visualization.

```python
# import libraries for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display settings
pd.set_option('display.max_columns', None)
```

```python
# load the dataset 
file_path = r"C:\Users\Dell\Downloads\Economics data analysis.csv"
df = pd.read_csv(file_path)

# Preview data
df.head()
```

```python
# view all column names
print(df.columns)

# get dataset information
df.info()

df.describe()

# check missing values in each column
df.isnull().sum()
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.replace(' ', '_')

# Remove duplicates
df = df.drop_duplicates()

# Display cleaned dataset
df.head()
```

```python
# sort data by cost of living in descending order and select top 10
top_expensive = df.sort_values(by='Cost_of_Living_Index', ascending=False).head(10)

top_expensive[['Country', 'Cost_of_Living_Index']]
```

```python
# sort data in ascending order to get cheapest countries
cheapest = df.sort_values(by='Cost_of_Living_Index', ascending=True).head(10)

cheapest[['Country', 'Cost_of_Living_Index']]
```

## Visualization Section

```python
# Bar chart- Expensive countries

plt.figure(figsize=(10,5))
plt.bar(top_expensive['Country'], top_expensive['Cost_of_Living_Index'])
plt.xticks(rotation=45)
plt.title('Top 10 Expensive Countries')
plt.show()
```

```python
# Scatter chart- Rent vs Cost

plt.scatter(df['Rent_Index'], df['Cost_of_Living_Index'])
plt.xlabel('Rent Index')
plt.ylabel('Cost of Living Index')
plt.title('Rent vs Cost of Living')
plt.show()
```

```python
# Scatter chart- Groceries vs Cost

plt.scatter(df['Groceries_Index'], df['Cost_of_Living_Index'])
plt.xlabel('Groceries Index')
plt.ylabel('Cost of Living Index')
plt.title('Groceries vs Cost')
plt.show()
```

```python
# Histogram

plt.hist(df['Cost_of_Living_Index'], bins=20)
plt.title('Distribution of Cost of Living')
plt.show()
```

```python
# Box Plot

df[['Cost_of_Living_Index','Rent_Index','Groceries_Index','Restaurant_Price_Index']].plot(kind='box')
plt.title('Distribution of Indexes')
plt.show()
```

```python
# Barh- Purchasing power

top_pp = df.sort_values(by='Local_Purchasing_Power_Index').tail(10)

plt.barh(top_pp['Country'], top_pp['Local_Purchasing_Power_Index'])
plt.title('Top Purchasing Power Countries')
plt.show()
```

```python
# Correlation Matrix

corr = df.corr(numeric_only=True)
print(corr)
```

```python
# Affordability Index
df['Affordability_Index'] = df['Local_Purchasing_Power_Index'] / df['Cost_of_Living_Index']

# Top affordable countries
top_affordable = df.sort_values(by='Affordability_Index', ascending=False).head(10)

# Plot
plt.barh(top_affordable['Country'], top_affordable['Affordability_Index'])
plt.title('Top Affordable Countries')
plt.show()
```

```python
# Heatmap using matplotlib
corr = df.corr(numeric_only=True)

plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title('Correlation Heatmap')
plt.show()
```

### Key Insights

#### High cost of living is strongly linked with high rent and groceries  
#### Purchasing power varies significantly across countries  
#### Some countries are expensive but still affordable due to high income

### Storytelling Insight

#### The analysis shows that cost of living alone does not determine how affordable a country is. Some countries with a high cost of living also have high purchasing power, making them relatively affordable for residents. 
#### On the other hand, countries with a low cost of living may still be less affordable due to low purchasing power. This highlights that affordability depends on both income levels and expenses, not just the cost of goods and services.

### Conclusion

#### The analysis shows that cost of living differs widely across countries.
#### Rent and groceries are the main contributors to high living costs.
#### Purchasing power is an important factor to measure real affordability.
#### Therefore, both cost and income levels must be considered for economic comparison.
